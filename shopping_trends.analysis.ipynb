```python
Gender_distribution = df['Gender'].value_counts()
Gender_distribution
```

```python
Gender_distribution = df['Gender'].value_counts()
Gender_distribution.plot(kind='pie', autopct='%1.1f%%', colors=['#66b3ff','#99ff99'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('count')
plt.show()


```

```python
 df2 = df.groupby('Gender')['Purchase Amount (USD)'].mean()
 df2
```

```python
df2.plot(kind='bar', figsize=(10, 6))
plt.title('Purchase Amount by Gender')
plt.xlabel('Gender')
plt.ylabel('Purchase Amount (USD)')
plt.show()
```

```python
df.boxplot(column='Purchase Amount (USD)', by='Gender')
plt.title('Purchase Amount by Gender')
plt.xlabel('Gender')
plt.ylabel('Purchase Amount (USD)')
plt.suptitle('') # Suppress the default title
plt.show()
```

```python
age_distribution = df['Age'].value_counts()
age_distribution
```

```python
df.groupby('Age')['Purchase Amount (USD)'].mean()
```

```python
item_purchased = df['Item Purchased'].value_counts().head(5)
item_purchased
```

```python
item_purchased.plot(kind='bar', figsize=(10, 6))
plt.title('Item Purchased')
plt.xlabel('Item Purchased')
```

```python
category_distribution = df['Category'].value_counts()
category_distribution
```

```python
category_distribution.plot(kind='bar', figsize=(10, 6))
plt.title('Category Distribution')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

```python
season_avg_purchase = df.groupby('Season')['Purchase Amount (USD)'].mean()
season_avg_purchase
```

```python
season_avg_purchase.plot(kind='bar', figsize=(10, 6))
plt.title('Average Purchase Amount by Season')
plt.xlabel('Season')
plt.ylabel('Average Purchase Amount (USD)')
plt.show()
```

```python
payment_method_distribution = df['Payment Method'].value_counts()
payment_method_distribution
```

```python
payment_method_distribution.plot(kind='pie', autopct='%1.1f%%', colors=['Blue','red'])
plt.title('Payment Method Distribution')
plt.xlabel('Payment Method')
plt.ylabel('count')
plt.show()
```

```python
rating_avg_by_subsscription = df.groupby('Subscription Status')['Review Rating'].mean()
rating_avg_by_subsscription
```

```python
rating_avg_by_subsscription.plot(kind='bar', figsize=(10, 6))
plt.title('Average Review Rating by Subscription Status')
plt.xlabel('Subscription Status')
plt.ylabel('Average Review Rating')
plt.show()
```

```python
discount_avg_by_purchase = df.groupby('Discount Applied')['Purchase Amount (USD)'].mean()
discount_avg_by_purchase
```

```python
discount_avg_by_purchase.plot(kind='bar', figsize=(10, 6))
plt.title('Average Purchase Amount by Discount Applied')
plt.xlabel('Discount Applied')
plt.ylabel('Average Purchase Amount (USD)')
plt.show()
```
